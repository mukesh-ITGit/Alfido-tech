```python
# ------------------------------
# Install dependencies first (run this in a Colab cell)
# ------------------------------
!pip install gradio groq pdfplumber nest_asyncio

import gradio as gr
import groq
import pdfplumber
import nest_asyncio
import os

# Apply nest_asyncio to avoid event loop errors in Colab
nest_asyncio.apply()

# ------------------------------
# API Key – from Colab secrets or environment
# ------------------------------
try:
    from google.colab import userdata
    GROQAPI_KEY = userdata.get('GROQ_API_KEY')
except ImportError:
    GROQAPI_KEY = os.getenv("GROQ_API_KEY")

client = groq.Groq(api_key=GROQAPI_KEY)

# Use a currently supported free model
MODEL_NAME = "llama-3.1-8b-instant"  # also try "gemma2-9b-it" if needed

# ------------------------------
def extract_text_from_pdf(file_path):
    text = ""
    with pdfplumber.open(file_path) as pdf:
        for page in pdf.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + "\n"
    return text.strip()

def generate_mcqs(context, num_questions=10):
    context = context[:3000]
    prompt = f"""Based on the following content, generate {num_questions} multiple-choice questions.
Each question should have 4 options (A, B, C, D) and indicate the correct answer.
Format the output as a numbered list exactly like this:
1. Question text
   A) Option A
   B) Option B
   C) Option C
   D) Option D
   Correct Answer: A

Content:
{context}
"""
    completion = client.chat.completions.create(
        model=MODEL_NAME,
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7,
        max_tokens=1500,
    )
    return completion.choices[0].message.content.strip()

def process_input(pdf_file, text_input):
    if pdf_file is not None:
        context = extract_text_from_pdf(pdf_file.name)
    elif text_input.strip():
        context = text_input.strip()
    else:
        return "Please upload a PDF or enter some text."
    if not context:
        return "No text could be extracted."
    try:
        mcqs = generate_mcqs(context)
        return mcqs
    except Exception as e:
        return f"Error: {str(e)}\n\nTip: Check supported free models at https://console.groq.com/docs/models"

# ------------------------------
with gr.Blocks(title="MCQ Generator (Groq)") as demo:
    gr.Markdown("# 📚 MCQ Generator with Groq Free API")
    gr.Markdown("Upload a PDF or paste text, then get 10 multiple‑choice questions.")

    with gr.Row():
        pdf_input = gr.File(label="Upload PDF", file_types=[".pdf"])
        text_input = gr.Textbox(label="Or paste text here", lines=5, placeholder="Paste your content...")

    generate_btn = gr.Button("Generate 10 MCQs")
    output = gr.Textbox(label="Generated Questions", lines=20)

    generate_btn.click(
        fn=process_input,
        inputs=[pdf_input, text_input],
        outputs=output
    )

if __name__ == "__main__":
    # share=True gives a public link (needed in Colab)
    demo.queue().launch(share=True, debug=False)
```
